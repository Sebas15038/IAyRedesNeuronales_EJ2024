# 1. Cargar líbrerias

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
```

# 2. Cargar Datos

```python
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
cancer.target_names
```

# 3. Preprocesamiento

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# 4. Selección y entrenamiento del modelo

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
train_accuracy = clf.score(X_train, y_train)
print(train_accuracy)
```

# 5. Prueba del modelo

```python
y_pred = clf.predict(X_test)
```

```python
print(y_pred[1])
print(y_test[1])
```

```python
print(y_pred[20])
print(y_test[20])
```

```python
print(y_pred[-1])
print(y_test[-1])
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
print("Classification Report: \n", classification_report(y_test, y_pred, target_names=cancer.target_names))
```

# Visualizacion del arbol de decisión

```python
plt.figure(figsize=(20, 15))
tree.plot_tree(clf, filled=True, feature_names=cancer.feature_names, class_names=cancer.target_names)
plt.title("Decision Tree for Breast Cancer Diagnostic")
plt.savefig("decision_tree.png", dpi=300)
plt.show()
```
